```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import csv
import re
```

```python
def scrape_coursera_courses(page_number):
    coursera_courses_page_html = requests.get(f"https://www.coursera.org/courses?page={page_number}").text
    
    html_code_soup = BeautifulSoup(coursera_courses_page_html, 'lxml')
    
    course_box = html_code_soup.find_all('div', class_="cds-ProductCard-gridCard")
    
    with open('coursera_data.csv', 'a', newline='', encoding='utf-8') as csvfile:
        fieldnames = ['Course Name', 'Company Offering', 'Skills Gained', 'Ratings', 'Number of Ratings']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        
        if csvfile.tell() == 0:
            writer.writeheader()

        for course in course_box:

            course_name = course.find('h3', class_="cds-CommonCard-title css-6ecy9b").text.strip()
            
            company_offering_that_course = course.find('p', class_="cds-ProductCard-partnerNames css-vac8rf").text.strip()
            
            skills_that_will_be_gained_element = course.find('div', class_="cds-CommonCard-bodyContent")
            
            skills_that_will_be_gained = skills_that_will_be_gained_element.text.strip() if skills_that_will_be_gained_element else ""
            
            course_ratings_element = course.find('p', class_="css-2xargn")
            
            course_ratings = course_ratings_element.text.strip() if course_ratings_element else ""
            
            skills_that_will_be_gained = skills_that_will_be_gained.replace("Skills you'll gain:", "").strip()

            
            course_metadata_elements = course.find_all('div', class_="product-reviews css-pn23ng")
            
            num_reviews = 0
            
            if course_metadata_elements:
                course_metadata = course_metadata_elements[-1].text.strip()

                num_reviews_match = re.search(r'(\d+)K? reviews?', course_metadata)
                
                if num_reviews_match:
                    num_reviews = int(num_reviews_match.group(1)) * 1000 if 'K' in num_reviews_match.group() else int(num_reviews_match.group(1))

            writer.writerow({
                'Course Name': course_name,
                'Company Offering': company_offering_that_course,
                'Skills Gained': skills_that_will_be_gained,
                'Ratings': course_ratings,
                'Number of Ratings': num_reviews
            })

    print(f"Data for web-page {page_number} has been written to csv file")
```

```python
for page_num in range(1, 85):  
    scrape_coursera_courses(page_num)
```
